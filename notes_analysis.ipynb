## Chart showing how long it took to resolve notes

```python
##plot by currentStatus (stacked, custom colors, with year annotations)
import matplotlib.pyplot as plt

# Define the desired order and colors
desired_order = [
    'CURRENTLY_RATED_HELPFUL',
    'CURRENTLY_RATED_NOT_HELPFUL',
    'NEEDS_MORE_RATINGS'
]
color_map = {
    'CURRENTLY_RATED_HELPFUL': '#32CD32',   # Green
    'CURRENTLY_RATED_NOT_HELPFUL': '#FF4500', # Red
    'NEEDS_MORE_RATINGS': '#A9A9A9'        # Grey
}

# Only include statuses present in your data
statuses = [s for s in desired_order if s in status['currentStatus'].unique()]
colors = [color_map[s] for s in statuses]
data = [status.loc[status['currentStatus'] == s, 'time_diff'].dropna() for s in statuses]

plt.figure(figsize=(10, 6))
plt.hist(
    data,
    bins=30,
    stacked=True,
    color=colors,
    label=statuses,
    alpha=0.8
)

# Add vertical lines for each year (365, 730, 1095, ...)
max_days = status['time_diff'].max()
for year in range(1, int(max_days // 365) + 2):
    plt.axvline(x=year * 365, color='black', linestyle='--', linewidth=1)
    plt.text(year * 365, plt.ylim()[1]*0.95, f'{year} yr', rotation=90, va='top', ha='right', color='black', fontsize=9)

plt.xlabel('Number of days')
plt.ylabel('Count of notes')
plt.suptitle('More than 90% of notes are stuck in limbo for over 100 days')
plt.title('Out of all notes classified under \'Needs More Ratings\', most remained in that status for months and even years')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.12))  # Move legend down
plt.tight_layout()
plt.savefig('nmr_chart_by_status_stacked_custom.svg', format='svg')
# plt.show()
```
